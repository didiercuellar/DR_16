```python
from astropy.io import fits 
from astropy.table import Table 
import numpy as np 
import pandas as pd
```

```python
path=''
```

```python
hdul = fits.open('DR16Q.fits')
print(hdul[1].columns)
datos=Table(hdul[1].data)
```

```python
print(datos)
RA=datos['RA']
print(RA)
DEC=datos['DEC']
print(DEC)
```

```python
np.savetxt(path+'RA.csv',RA, fmt='%5.5f', header='RA')
```

```python
np.savetxt(path+'DEC.csv',DEC, fmt='%5.5f', header='DEC')
```
